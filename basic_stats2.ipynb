```python
import numpy as np
from scipy import stats

# Durability data (in millions of characters)
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
data
```

```python
n = len(data)               # sample size
x_bar = np.mean(data)       # sample mean
s = np.std(data) 
```

```python
confidence = 0.99
alpha = 1 - confidence
df = n - 1

# t critical value
t_crit = stats.t.ppf(1 - alpha/2, df)

# margin of error
me = t_crit * (s / np.sqrt(n))

# confidence interval
ci_t = (x_bar - me, x_bar + me)
print("99% CI using sample standard deviation (t-distribution):", ci_t)
```

```python
sigma = 0.2

# z critical value
z_crit = stats.norm.ppf(1 - alpha/2)

# margin of error
me_z = z_crit * (sigma / np.sqrt(n))

# confidence interval
ci_z = (x_bar - me_z, x_bar + me_z)
print("99% CI using known population standard deviation (z-distribution):", ci_z)
```
